# 요일별로 인덱스 추출하기

```python
from tqdm import tqdm

sun_index = []
mon_index = []
tue_index = []
wed_index = []

for i in tqdm(range(len(my_gpd2['ymd']))):
    if my_gpd2['ymd'].iloc[i] == '20180701':
        sun_index.append(i)
    elif my_gpd2['ymd'].iloc[i] == '20180702':
        mon_index.append(i)
    elif my_gpd2['ymd'].iloc[i] == '20180703':
        tue_index.append(i)
    elif my_gpd2['ymd'].iloc[i] == '20180704':
        wed_index.append(i)
```

## 월요일 출근시간만 그려보기 (07~09시)

```python
mon_df = my_gpd2.loc[mon_index]
```

```python
mon_df
```

```python
mon_columns = list(mon_df.columns)
```

```python
mon_columns.remove('t07_08')
mon_columns.remove('t08_09')
mon_columns.remove('emd_cd')
```

```python
mon_columns
```

```python
mon_df_2 = mon_df
```

```python
for i in range(len(mon_columns)):
    del mon_df_2[mon_columns[i]]
```

```python
mon_df_2['t07_09'] = mon_df_2['t07_08'] + mon_df_2['t08_09']
```

```python
del mon_df_2['t08_09']
del mon_df_2['t07_08']
```

# 승 하차 버스 정류장 정보 가져오기

```python
sungcha_07_09 = pd.read_csv("../7-9_sungcha.csv")
hacha_07_09 = pd.read_csv("../7-9_hacha.csv")
```

```python
bus_stop_list = list(set(list(sungcha_07_09['표준정류장ID'])))
bus_stop_list2 = list(set(list(hacha_07_09['표준정류장ID'])))
```

```python
hacha_07_09
```

```python
sungcha_index = []

for i in tqdm(range(len(bus_stop_list))):
    for j in range(len(sungcha_07_09['표준정류장ID'])):
        if sungcha_07_09['표준정류장ID'].iloc[j] == bus_stop_list[i]:
            sungcha_index.append(j)
            break
```

```python
hacha_index = []

for i in tqdm(range(len(bus_stop_list2))):
    for j in range(len(hacha_07_09['표준정류장ID'])):
        if hacha_07_09['표준정류장ID'].iloc[j] == bus_stop_list2[i]:
            hacha_index.append(j)
            break
```

```python
sungcha_07_09_2 = sungcha_07_09.loc[sungcha_index]
hacha_07_09_2 = hacha_07_09.loc[hacha_index]
```

# 출근시간 승하차 정류장, 학교 folium 시각화

```python
m = folium.Map(location=[latitude, longitude], zoom_start=10)#, tiles='color_map')

folium.GeoJson(
    converted_json,
).add_to(m)


for i in range(len(my_json2)):
    m.choropleth(
        geo_data = my_json2[i],
        name='인구수',
        data=mon_df_2,
        columns=['emd_cd', 't07_09'],
        key_on='properties.emd_cd',
        fill_color='YlGn',
        fill_opacity=0.7,
        line_opacity=0.3,
        color = 'gray',
        overlays=True
    )


for i in tqdm(range(0, len(sungcha_07_09_2['위도']))):
    folium.CircleMarker([sungcha_07_09_2['위도'].iloc[i], sungcha_07_09_2['경도'].iloc[i]], popup=('승차'+sungcha_07_09_2['정류소명']),radius=6,color='yellow',fill_color='yellow').add_to(m)


for i in tqdm(range(0, len(hacha_07_09_2['위도']))):
    folium.CircleMarker([hacha_07_09_2['위도'].iloc[i], hacha_07_09_2['경도'].iloc[i]], popup=('승차'+hacha_07_09_2['정류소명']),radius=4,color='green',fill_color='green').add_to(m)

schools = schools.fillna(0.0)

for i in tqdm(range(0, len(schools['latitude']))):
    folium.CircleMarker([schools['latitude'].iloc[i], schools['longitude'].iloc[i]], popup=schools['학교명'].iloc[i],radius=4,color='blue',fill_color='blue').add_to(m)

```

```python
m.save('youdong_07_09_mon_school_and_stop.html')
```
